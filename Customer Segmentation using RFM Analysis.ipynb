### REM Analysis

```python
import datetime as dt
```

```python
retail_df["InvoiceDate"].max()
```

```python
today_date=dt.datetime(2011, 12, 11)
```

```python
rfm = retail_df.groupby('CustomerID').agg({'InvoiceDate': lambda date: (today_date - date.max()).days,
                                         'InvoiceNo'  : lambda num:   num.nunique(),
                                         'TotalPrice' : lambda TotalPrice: TotalPrice.sum()})

rfm
```

```python
rfm.columns=["Recency","Ferquency","Monetary"]

rfm
```

```python
rfm = rfm[rfm["Monetary"]>0]

rfm
```

```python
rfm.describe().T
```

```python
rfm["recency_score"] = pd.qcut(rfm["Recency"], 5, labels=[5, 4, 3, 2, 1])

rfm
```

```python
rfm["ferquency_score"] = pd.qcut(rfm["Ferquency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5])

rfm
```

```python
rfm["monetary_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1, 2, 3, 4, 5])

rfm
```

```python
rfm.sample(10)
```

```python
rfm["RFM_SCORE"]=(rfm["recency_score"].astype(str)+rfm["ferquency_score"].astype(str))
```

```python
rfm.head()
```

```python
#segmenting Customers Using RFM Score

seg_map = {
    r'[1-2][1-2]':'hibernating',
    r'[1-2][3-4]':'at_Risk',
    r'[1-2]5':'cant_loose',
    r'3[1-2]':'about_to_sleep',
    r'33':'need_attention',
    r'[3-4][4-5]':'loyal_customers',
    r'41':'promising',
    r'51':'new_customers',
    r'[4-5][2-3]':'potential_loyalists',
    r'5[4-5]':'champions'
}

rfm['segment'] = rfm['RFM_SCORE'].replace(seg_map, regex=True)
rfm.head()
```

```python
rfm[["segment","Recency","Ferquency","Monetary"]].groupby("segment").agg(["mean","count","max"]).round()
```
